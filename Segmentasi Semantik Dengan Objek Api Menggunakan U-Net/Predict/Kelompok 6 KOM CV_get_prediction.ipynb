```python
## Imports
import os
import sys
import random

import numpy as np
import cv2
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras

## Seeding 
seed = 42
random.seed = seed
np.random.seed = seed
tf.seed = seed
```

# Test for Scoring

```python


def get_prediction(PATH):
    image = cv2.imread(PATH)
    image = cv2.resize(image, (128,128))
    x = np.expand_dims(image, axis=0)
    img = x.astype('float32')/256

    model_unet = keras.models.load_model("unet (save model) spot the fire 20 epochs.h5")

    #classify fire/non fire with cnn
    model_classifier = keras.models.load_model("fire_detection.h5")
    pred = model_classifier.predict(np.expand_dims(img,axis=0))[0]
    if (np.argmax(pred) == 1):
        result = model_unet.predict(x)
        result = result > 0.5
    else:
        result = np.zeros((1,128,128,1))
        
    fig = plt.figure()
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    ax = fig.add_subplot(1, 2, 1)
    ax.imshow(image)
    ax = fig.add_subplot(1, 2, 2)
    ax.imshow(np.reshape(result[0]*255, (128, 128)), cmap="gray")
```
