Text:

*   A = "my man is not usually this ugly"
*   B = "my woman is not ugly and she is big"
*   C = "that man is ugly and very big



```python
A =  "my man is not usually this ugly"
B = "my woman is not ugly and she is big"
C = "that man is ugly and very big"
```

```python
import pandas as pd
```

```python
def TFs(wordDict, tokens):
    tfDict = {}
    wCount = len(tokens)
    for word, count in wordDict.items():
        tfDict[word] = count / float(wCount)
    return tfDict
```

```python
def token(doc):
  word = doc.split(' ')
  return word
```

```python
tA = token(A)
tB = token(B)
tC = token(C)
```

```python
unique1 = set(tA).union(set(tB))
unique = set(unique1).union(set(tC))
unique
```

```python
numA = dict.fromkeys(unique, 0)
for word in tA:
    numA[word] += 1

numB = dict.fromkeys(unique, 0)
for word in tB:
    numB[word] += 1

numC = dict.fromkeys(unique, 0)
for word in tC:
    numC[word] += 1
```

```python
df = pd.DataFrame([numA, numB, numC])
df
```

```python
tfA = TFs(numA, tA)
tfB = TFs(numB, tB)
tfC = TFs(numC, tC)
```

```python
dfs = pd.DataFrame([tfA, tfB, tfC])
dfs
```

```python
import math
def IDF(doc):
    N = len(doc)
    idfDict = dict.fromkeys(doc[0].keys(), 0)
    for docs in doc:
        for word, val in docs.items():
            if val > 0:
                idfDict[word] += 1

    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))
    return idfDict
```

```python
idf = IDF([numA, numB, numC])
idf_df = pd.DataFrame([idf])
idf_df
```

```python
def TFIDF(tokens, idf):
    tfidf = {}
    for word, val in tokens.items():
        tfidf[word] = val * idf[word]
    return tfidf
```

```python
tfidfA = TFIDF(tfA, idf)
tfidfB = TFIDF(tfB, idf)
tfidfC = TFIDF(tfC, idf)
df = pd.DataFrame([tfidfA, tfidfB, tfidfC])
df
```
